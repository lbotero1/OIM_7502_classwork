### 21. Determine and comment on the day of the week that movies are released. What is notable about release days and months throughout the year? (Use pd.pivot_table)

```python
release['date'] = pd.to_datetime(release['date'])
release['day_of_week'] = release['date'].dt.day_name()
release['month'] = release['date'].dt.month_name()
release_summary = pd.pivot_table(release, index='day_of_week', columns='month', aggfunc='size', fill_value=0)
print(release_summary)
```

Comment: Fridays have the highest number of movie releases across all months, indicating a trend in the industry to release movies on Fridays, possibly to attract larger audiences over the weekend.
Thursdays also have a significant number of movie releases, being the second largest day off the week with movie releases, particularly in months like March, April, and November.Saturdays and Sundays generally have fewer movie releases compared to weekdays, Sunday being the day with fewer movies released across all months.
